```python
# Cell 1: Импорты и пути к файлам
%matplotlib inline

import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, clear_output
import ipywidgets as widgets
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_absolute_error

# Пути к Excel-файлам (они должны лежать рядом с этим .ipynb)
file_sate        = 'Populatia_sate_comune_orase_2014-2024 (1).xlsx'
file_pop031600   = 'pop031600_20250401-013435.xlsx'
file_pop070100_a = 'pop070100_20250401-013609.xlsx'
file_pop070100_b = 'pop070100_20250401-013641.xlsx'
file_nr_res      = 'Nr_pop_resedinta_obisnuita_profil_teritorial_01_01_2014_2023.xlsx'

# Проверка
print("Рабочая папка:", os.getcwd())
print("Файлы в ней:", os.listdir())
```

```python
# Cell 2: Загрузка всех наборов данных
# 2.1 – «sate» содержит по листам данные за 2014–2024 гг.
df_sate_dict = pd.read_excel(
    file_sate,
    header=4,        # заголовки на 5-й строке (index=4)
    sheet_name=None  # читаем все листы
)
# 2.2 – остальные файлы с одним листом, заголовок тоже на строке 5
df_pop031600   = pd.read_excel(file_pop031600,   header=4)
df_pop070100_a = pd.read_excel(file_pop070100_a, header=4)
df_pop070100_b = pd.read_excel(file_pop070100_b, header=4)
df_nr_res      = pd.read_excel(file_nr_res,      header=4)

# Быстрый контроль размеров
print("sate (по листам):", {yr: df.shape for yr, df in df_sate_dict.items()})
print("pop031600:",   df_pop031600.shape)
print("pop070100_a:", df_pop070100_a.shape)
print("pop070100_b:", df_pop070100_b.shape)
print("nr_res:",      df_nr_res.shape)
```

```python
#3 === Обновлённая ячейка: обработка и очистка данных ===

frames = []

for year, df in df_sate_dict.items():
    df = df.copy()
    df.columns = ['Region', 'Code', 'Locality', 'Total', 'Male', 'Female'] + list(df.columns[6:])
    df = df[df['Region'].str.lower().str.strip() != 'total pe tara']
    df[['Total', 'Male', 'Female']] = df[['Total', 'Male', 'Female']].apply(
        pd.to_numeric, errors='coerce'
    )
    df['Year'] = int(year)
    frames.append(df[['Region', 'Code', 'Locality', 'Year', 'Total', 'Male', 'Female']])

df_sate_all = pd.concat(frames, ignore_index=True)

# ✅ Удаляем дубликаты (оставляем по 1 записи на Code+Year)
df_sate_all = df_sate_all.drop_duplicates(subset=['Code', 'Year'])

print("Объединённый размер df_sate_all:", df_sate_all.shape)
df_sate_all.head()

```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score, mean_absolute_error
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display, clear_output

# 1) Уникальные регионы
regions = sorted(df_sate_all['Region'].dropna().astype(str).unique())

# 2) Виджеты
region_dd   = widgets.Dropdown(options=regions, description='Регион:')
locality_dd = widgets.Dropdown(description='Населённый:')
factor_dd   = widgets.Dropdown(options=['Total', 'Male', 'Female'], description='Фактор:')
run_btn     = widgets.Button(description='Анализ')
out         = widgets.Output()

# 3) Обновление списка Locality (Locality + Code)
def on_region_change(change):
    df_filtered = df_sate_all[df_sate_all['Region'].astype(str) == change['new']]
    df_unique = df_filtered[['Code', 'Locality']].drop_duplicates()
    locality_dd.options = [
    f"{row.Locality} ({int(row.Code)})" for _, row in df_unique.iterrows()
    ]


region_dd.observe(on_region_change, names='value')
region_dd.value = regions[0]  # Инициализация

# 4) Нажатие на кнопку анализа
def on_run(_):
    with out:
        clear_output()
        reg = region_dd.value
        loc_code_raw = locality_dd.value
        fac = factor_dd.value

        # Извлекаем код из строки вида "DROCHIA (36010)"
        if '(' in loc_code_raw:
            code = int(loc_code_raw.split('(')[-1].rstrip(')'))
        else:
            print("Ошибка: не удалось определить код населённого пункта.")
            return

        df_sel = df_sate_all[
            (df_sate_all['Code'] == code)
        ]

        if df_sel.empty:
            print("Нет данных для выбранного пункта.")
            return

        # Подготовка данных
        df_sel = df_sel.sort_values('Year')
        X = df_sel[['Year']].values
        y = df_sel[fac].values

        # Линейная модель
        model_lin = LinearRegression().fit(X, y)
        y_pred_lin = model_lin.predict(X)
        r2_lin = r2_score(y, y_pred_lin)
        mae_lin = mean_absolute_error(y, y_pred_lin)

        # Если слабая — полиномиальная регрессия
        if r2_lin < 0.2:
            poly = PolynomialFeatures(degree=2)
            X_poly = poly.fit_transform(X)
            model_poly = LinearRegression().fit(X_poly, y)
            y_pred_poly = model_poly.predict(X_poly)
            r2 = r2_score(y, y_pred_poly)
            mae = mean_absolute_error(y, y_pred_poly)
            model_type = "Полиномиальная регрессия (2 степени)"
        else:
            y_pred_poly = y_pred_lin
            r2 = r2_lin
            mae = mae_lin
            model_type = "Линейная регрессия"

        # Вывод метрик
        print(f"R² = {r2:.3f}, MAE = {mae:.1f}")
        if r2 < 0.2:
            print("⚠️ Внимание: данные слабо поддаются линейному анализу, используется полиномиальная модель.")

        # График
        plt.figure(figsize=(6,4))
        plt.scatter(df_sel['Year'], y, label='Фактические', alpha=0.7)
        plt.plot(df_sel['Year'], y_pred_poly, 'r-', label=model_type)
        plt.xlabel('Год'); plt.ylabel(fac)
        plt.title(f"{loc_code_raw} ({reg})")
        plt.legend()
        plt.grid(True)
        plt.show()

# 5) Подключаем кнопку
run_btn.on_click(on_run)

# 6) Отображаем интерфейс
display(region_dd, locality_dd, factor_dd, run_btn, out)
```

```python
# Cell 5: Глобальная тенденция – суммарная численность по годам
tot_year = df_sate_all.groupby('Year')['Total'].sum().sort_index()
plt.figure(figsize=(8,4))
plt.plot(tot_year.index, tot_year.values, marker='o')
plt.xlabel('Год'); plt.ylabel('Население (Total)')
plt.title('Общая численность населения Молдовы (2014–2024)')
plt.grid(True)
plt.show()
```

```python
# Cell 6: Корреляционная матрица Year/Total/Male/Female
cols = ['Year','Total','Male','Female']
corr = df_sate_all[cols].corr()
fig, ax = plt.subplots(figsize=(4,4))
im = ax.imshow(corr, interpolation='nearest')
ax.set_xticks(range(len(cols))); ax.set_yticks(range(len(cols)))
ax.set_xticklabels(cols, rotation=45, ha='right')
ax.set_yticklabels(cols)
for i in range(len(cols)):
    for j in range(len(cols)):
        ax.text(j, i, f"{corr.iloc[i,j]:.2f}",
                ha='center', va='center', color='white')
fig.colorbar(im, ax=ax)
plt.title('Корреляция показателей')
plt.tight_layout()
plt.show()
```

```python
# Cell 7: Топ-10 населённых пунктов по приросту населения 2014→2024
df_2014 = df_sate_all[df_sate_all.Year==2014]
df_2024 = df_sate_all[df_sate_all.Year==2024]
# Объединяем и считаем прирост
growth = (df_2024.set_index(['Region','Locality'])['Total'] -
          df_2014.set_index(['Region','Locality'])['Total']).dropna()
top10 = growth.sort_values(ascending=False).head(10)
# Бар-чарт
plt.figure(figsize=(8,5))
plt.barh(range(len(top10)), top10.values, color='skyblue')
plt.yticks(range(len(top10)), [f"{r} – {l}" for r,l in top10.index])
plt.gca().invert_yaxis()
plt.xlabel('Прирост населения')
plt.title('Топ-10 мест с наибольшим приростом (2014→2024)')
plt.tight_layout()
plt.show()
```

```python
# Cell 8: Подготовка combined без dropna() и выравнивание годов
import pandas as pd

def get_year_cols(df):
    year_cols, year_map = [], {}
    for c in df.columns:
        if isinstance(c, int) and 1900 <= c <= 2100:
            year_cols.append(c); year_map[c] = c
        elif isinstance(c, str) and c.isdigit():
            y = int(c)
            if 1900 <= y <= 2100:
                year_cols.append(c); year_map[c] = y
    return year_cols, year_map

# imm
year_cols_a, year_map_a = get_year_cols(df_pop070100_a)
id_vars_a = [c for c in df_pop070100_a.columns if c not in year_cols_a]
imm = df_pop070100_a.melt(id_vars=id_vars_a, value_vars=year_cols_a,
                          var_name='Year', value_name='Immigrants')
imm['Year'] = imm['Year'].map(year_map_a)
imm['Immigrants'] = pd.to_numeric(imm['Immigrants'], errors='coerce').fillna(0)
imm_year = imm.groupby('Year')['Immigrants'].sum().sort_index()

# emm
year_cols_b, year_map_b = get_year_cols(df_pop070100_b)
id_vars_b = [c for c in df_pop070100_b.columns if c not in year_cols_b]
emm = df_pop070100_b.melt(id_vars=id_vars_b, value_vars=year_cols_b,
                          var_name='Year', value_name='Emigrants')
emm['Year'] = emm['Year'].map(year_map_b)
emm['Emigrants'] = pd.to_numeric(emm['Emigrants'], errors='coerce').fillna(0)
emm_year = emm.groupby('Year')['Emigrants'].sum().sort_index()

mig_year = imm_year.subtract(emm_year, fill_value=0).sort_index()

# tot
tot = df_sate_all.groupby('Year')['Total'].sum().sort_index()

# Выравниваем миграцию по тем же годам, что и tot
mig_aligned = mig_year.reindex(tot.index, fill_value=0)

combined = pd.DataFrame({
    'Total': tot,
    'NetMigration': mig_aligned
})
# делаем DatetimeIndex
combined.index = pd.to_datetime(combined.index.astype(str) + '-01-01')

combined.head()
```

```python
# Cell X: Прогноз до 2035 с ExponentialSmoothing (мультипликативный тренд)

import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.holtwinters import ExponentialSmoothing

# 1) Сколько шагов до 2035
last_year = combined.index.year.max()      # 2024
steps     = 2035 - last_year               # например, 11

# 2) Исторический ряд
hist = combined['Total']

# Явно задаём параметры сглаживания, чтобы optimization=False
from statsmodels.tsa.holtwinters import ExponentialSmoothing

model_hw = ExponentialSmoothing(
    hist, 
    trend='mul', 
    seasonal=None, 
    freq='YS'
).fit(
    optimized=False,       # не пытаться оптимизировать
    smoothing_level=0.8,   # α
    smoothing_trend=0.2    # β
)

forecast_hw = model_hw.forecast(steps)


# 4) Делаем прогноз
forecast_hw = model_hw.forecast(steps)

# 5) Рисуем историю и прогноз до 2035
future_idx = pd.date_range(
    start=pd.to_datetime(f"{last_year+1}-01-01"),
    periods=steps,
    freq='YS'
)

plt.figure(figsize=(10,5))
plt.plot(hist.index, hist, 'o-', label='История (2014–2024)')
plt.plot(future_idx, forecast_hw, 'r--', label='Holt Winterns (mul) до 2035')
plt.xlabel('Год')
plt.ylabel('Население')
plt.title('Прогноз населения Молдовы до 2035 (Holt’s mul trend)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Cell X: Улучшенные графики — Топ-5 регионов, тепловая карта (лог), структура пола

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# График: Топ-5 регионов по населению (масштаб Y от 50 000)
# Топ-5 регионов
region_totals = df_sate_all.groupby(['Region', 'Year'])['Total'].sum().reset_index()
top_regions = df_sate_all.groupby('Region')['Total'].sum().nlargest(5).index
filtered = region_totals[region_totals['Region'].isin(top_regions)]

# Построение графика
plt.figure(figsize=(10, 6))
for region in top_regions:
    data = filtered[filtered['Region'] == region]
    plt.plot(data['Year'], data['Total'], label=region)

plt.title("Динамика численности населения (Топ-5 регионов)")
plt.xlabel("Год")
plt.ylabel("Общее население")

# ✅ Устанавливаем шаг оси Y в 50 000
y_max = filtered['Total'].max()
plt.yticks(np.arange(0, y_max + 30000, 30000))

plt.grid(True)
plt.legend(loc='upper right')
plt.tight_layout()
plt.show()


# 2. График: Тепловая карта логарифма численности населения
# Cell X: Тепловая карта процентного изменения населения по регионам и годам

# Создаём сводную таблицу населения
heatmap_data = df_sate_all.pivot_table(
    index='Region',
    columns='Year',
    values='Total',
    aggfunc='sum'
)

# Вычисляем процентное изменение по сравнению с предыдущим годом
percent_change = heatmap_data.pct_change(axis=1) * 100  # в процентах
percent_change = percent_change.drop(columns=2014)

# Визуализация
plt.figure(figsize=(14, 8))
sns.heatmap(
    percent_change,
    cmap="coolwarm",
    center=0,
    annot=True,
    fmt='.1f',
    vmin=-5, vmax=5
)
plt.title("Процентное изменение численности населения по регионам (год к году, %)")
plt.xlabel("Год")
plt.ylabel("Регион")
plt.tight_layout()
plt.show()


# 3. График: Абсолютное количество мужчин и женщин (stacked area)
gender_df = df_sate_all.groupby('Year')[['Male', 'Female']].sum()
years = gender_df.index
male_vals = gender_df['Male'].values
female_vals = gender_df['Female'].values

plt.figure(figsize=(10, 6))
plt.stackplot(years, male_vals, female_vals, labels=['Мужчины', 'Женщины'],
              colors=['#1f77b4', '#ff7f0e'])
plt.title("Абсолютное число мужчин и женщин в населении (2014–2024)")
plt.xlabel("Год")
plt.ylabel("Численность")
plt.legend(loc='upper right')
plt.grid(True)
plt.tight_layout()
plt.show()
```
